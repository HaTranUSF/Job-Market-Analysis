Group job titles to primary roles

```python

# Mapping dictionary (from top 20 titles → grouped roles)
mapping = {
    # Data Engineer
    "senior data engineer": "Data Engineer",
    "data engineer": "Data Engineer",
    "lead data engineer": "Data Engineer",
    "manager, data engineering": "Data Engineer",
    "senior data engineer, public company": "Data Engineer",

    # Data Analyst
    "senior data analyst": "Data Analyst",
    "data analyst": "Data Analyst",

    # Data Scientist
    "data scientist": "Data Scientist",
    "senior data scientist": "Data Scientist",
    "lead data scientist": "Data Scientist",

    # Machine Learning Engineer
    "staff machine learning engineer, series a": "Machine Learning Engineer",
    "senior machine learning engineer": "Machine Learning Engineer",
    "machine learning engineer": "Machine Learning Engineer",
    "senior machine learning engineer - ai": "Machine Learning Engineer",
    "senior mls engineer, autonomous driving startup": "Machine Learning Engineer",

    # MLOps
    "senior mlops engineer": "MLOps Engineer",

    # Data Architect / Database
    "data architect": "Data Architect / Database",
    "senior database administrator": "Data Architect / Database",

    # Data Security
    "manager, data loss prevention (dlp) engineer (symantec)": "Data Security Engineer",
    "principal associate, data loss prevention (dlp) engineer": "Data Security Engineer"
}

# Function to standardize job titles into groups
def map_job_title(title):
    title_lower = title.lower()
    for key, value in mapping.items():
        if key in title_lower:   # case-insensitive matching
            return value
    return "Other"  # default for all titles not in top 20

# Apply mapping
df["job_group"] = df["job_title"].apply(map_job_title)

print(df)
```

Cleansed and standardized a column of comma-separated job skills, identified the top 10 most frequent skills, and created one-hot encoded columns for those skills in the original dataframe

```python

# 1. Ensure list of skills is ready
df_1["job_skills_list"] = df_1["job_skills"].fillna("").str.split(",").apply(
    lambda x: [s.strip().lower() for s in x if s.strip() != ""]
)

# 2. Explode into long format
df_long = df_1.explode("job_skills_list").reset_index()
df_long.rename(columns={"index": "row_id"}, inplace=True)

# 3. Count frequencies
skills_count = df_long["job_skills_list"].value_counts()
top_10_skills = skills_count.head(10).index.tolist()
# Another way: top_10_skills = skills_count.index [:10].tolist()

# 4. Create dummy (one-hot) columns only for top 10
for skill in top_10_skills:
    df_1[skill] = df_1["job_skills_list"].apply(lambda skills: int(skill in skills))

print(df_1.head())
```
